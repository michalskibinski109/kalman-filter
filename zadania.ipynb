```python
from kalman_filter import *
import numpy as np
import matplotlib.pyplot as plt
```

### zadania:
utworzenie punktu poruszającego się w jednym wymiarze a następnie implementacja filtrów:
* $\alpha$
* $\alpha - \beta$
* $\alpha - \beta - \gamma$
* filtru kalmana dla 1 wymiaru


```python
p = Point(np.random.randint(-5,5), np.random.randint(-5,5))
p(20)
a = p.history
f = OneDimFilters(a)
x = f.alphaFilter()
y = f.alphaBetaFilter()
z = f.alphaBetaGammaFilter()
g = f.kalmanFilter(r = .1, std = .2, q = .1)
N = list(range(0, len(a)))
plt.figure(figsize=(10,10))
plt.plot(N, a, marker = "^")
plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.plot(g, marker = ".")
plt.legend(['true values', 'alpha predictet', 'alpha - beta ', 'a-b-g', 'kalman'])
plt.show()
```
